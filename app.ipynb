```python
import requests
import pandas as pd
url = 'https://etmarketsapis.indiatimes.com/ET_TechnicalScreeners/getFilteredData'
stocktype, output = [], []
payload = {
    'exchangeId': 50,
    'fieldNames': "symbol",
    'pageNumber': 1,
    'pageSize': 1000,
    'predefinedFilterName': "WHITE_SPINNING_TOP",
    'sortedField': "performanceD1",
    'sortedOrder': "desc"
}
r = requests.post(url=url, json=payload)
data = r.json()['page']
for item in data:
  output.append(item['symbol'])
  stocktype.append('WHITE SPINNING TOP')

payload = {
    'exchangeId': 50,
    'fieldNames': "symbol",
    'pageNumber': 1,
    'pageSize': 1000,
    'predefinedFilterName': "BLACK_SPINNING_TOP",
    'sortedField': "performanceD1",
    'sortedOrder': "desc"
}
r = requests.post(url=url, json=payload)
data = r.json()['page']
for item in data:
  output.append(item['symbol'])
  stocktype.append('BLACK SPINNING TOP')
df = pd.DataFrame({'Symbol': output, 'Type': stocktype})
display(df)
```

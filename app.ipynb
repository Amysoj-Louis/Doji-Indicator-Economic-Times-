```python
# Importing required libraries
import requests
import pandas as pd

# API endpoint for filtered data
url = 'https://etmarketsapis.indiatimes.com/ET_TechnicalScreeners/getFilteredData'

# Initializing empty lists to store stock symbols and stock types
stocktype, output = [], []

# Payload for filter 'WHITE SPINNING TOP'
payload = {
    'exchangeId': 50,  # Exchange ID
    'fieldNames': "symbol",  # Fields to retrieve
    'pageNumber': 1,  # Page number
    'pageSize': 1000,  # Page size
    'predefinedFilterName': "WHITE_SPINNING_TOP",  # Filter name
    'sortedField': "performanceD1",  # Sorting field
    'sortedOrder': "desc"  # Sorting order
}

# Making a POST request to the API endpoint
r = requests.post(url=url, json=payload)

# Extracting data from the API response
data = r.json()['page']

# Iterating through the data and appending symbols and stock types to the respective lists
for item in data:
  output.append(item['symbol'])
  stocktype.append('WHITE SPINNING TOP')

# Payload for filter 'BLACK SPINNING TOP'
payload = {
    'exchangeId': 50,  # Exchange ID
    'fieldNames': "symbol",  # Fields to retrieve
    'pageNumber': 1,  # Page number
    'pageSize': 1000,  # Page size
    'predefinedFilterName': "BLACK_SPINNING_TOP",  # Filter name
    'sortedField': "performanceD1",  # Sorting field
    'sortedOrder': "desc"  # Sorting order
}

# Making a POST request to the API endpoint
r = requests.post(url=url, json=payload)

# Extracting data from the API response
data = r.json()['page']

# Iterating through the data and appending symbols and stock types to the respective lists
for item in data:
  output.append(item['symbol'])
  stocktype.append('BLACK SPINNING TOP')

# Creating a Pandas dataframe from the lists
df = pd.DataFrame({'Symbol': output, 'Type': stocktype})

# Displaying the dataframe
display(df)
```
